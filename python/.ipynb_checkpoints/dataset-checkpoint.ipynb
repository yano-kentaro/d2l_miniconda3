```python
#=================================================================|1
#							データセットの作成
#========================================================|2021.12.19
# v1: 塊に別れたデータセットの作成

from sklearn.datasets import make_blobs
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

X, y = make_blobs(
	random_state = 3, # シード値
	n_features = 2, # 特徴量
	centers = 3, # 塊数
	cluster_std = 1, # ばらつき
	n_samples = 300, # サンプル数
)

# 特徴量(X)でデータフレームを作成
df = pd.DataFrame(X)

# 分類(y)をtargetの列として追加
df["target"] = y

# 各々のデータフレームに分類
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]

# 散布図として描画
plt.figure(figsize=(5,5))
plt.scatter(df0[0],df0[1],color="b",alpha=0.5)
plt.scatter(df1[0],df1[1],color="r",alpha=0.5)
plt.scatter(df2[0],df2[1],color="g",alpha=0.5)
plt.grid()
plt.show()
```
