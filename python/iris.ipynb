```python
%matplotlib inline
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

# sklearnでアヤメのデータセットを変数に格納
iris = datasets.load_iris()

# pandasでアヤメのデータセットをデータフレームに格納
df = pd.DataFrame(iris.data)

# データフレームのカラム名を変更
df.columns = iris.feature_names
# アヤメの品種を表示
df["target"] = iris.target

#=================================================================1
#							二つの特徴量から散布図を作成
#========================================================2021.12.19
# v1: 散布図の描画

# グラフサイズ設定
plt.figure(figsize=(5,5))

# アヤメの品種次第で格納するデータフレームを分岐
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]

# ラベル名の設定
xlabel_name = "sepal width (cm)"
ylabel_name = "sepal length (cm)"
# 散布図設定
plt.scatter(df0[xlabel_name],df0[ylabel_name],color="b",alpha=0.5)
plt.scatter(df1[xlabel_name],df1[ylabel_name],color="r",alpha=0.5)
plt.scatter(df2[xlabel_name],df2[ylabel_name],color="g",alpha=0.5)
plt.xlabel(xlabel_name)
plt.ylabel(ylabel_name)
plt.grid()
plt.show()
```

```python
%matplotlib inline
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# sklearnでアヤメのデータセットを変数に格納
iris = datasets.load_iris()

# pandasでアヤメのデータセットをデータフレームに格納
df = pd.DataFrame(iris.data)

# データフレームのカラム名を変更
df.columns = iris.feature_names
# アヤメの品種を表示
df["target"] = iris.target
# アヤメの品種次第で格納するデータフレームを分岐
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]

#=================================================================1
#							三つの特徴量から散布図を作成
#========================================================2021.12.19
# v1: 3D散布図の描画

# グラフサイズ設定
ax = Axes3D(plt.figure(figsize=(5,5)))

# ラベル名の設定
xx = "sepal width (cm)"
yy = "sepal length (cm)"
zz = "petal length (cm)"
# 散布図設定
ax.scatter(df0[xx],df0[yy],df0[zz],color="b")
ax.scatter(df1[xx],df1[yy],df1[zz],color="r")
ax.scatter(df2[xx],df2[yy],df2[zz],color="g")
ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
# 視点を変更
ax.view_init(90,270)
plt.show()
```
