```python
%matplotlib inline
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

# sklearnでアヤメのデータセットを変数に格納
iris = datasets.load_iris()

# pandasでアヤメのデータセットをデータフレームに格納
df = pd.DataFrame(iris.data)

# データフレームのカラム名を変更
df.columns = iris.feature_names
# アヤメの品種を表示
df["target"] = iris.target

#=================================================================|1
#												一つの特徴量からヒストグラムを作成
#========================================================|2021.12.19
# v1: ヒストグラムを描画

# グラフサイズ設定
plt.figure(figsize=(5,5))

# アヤメの品種次第で格納するデータフレームを分岐
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]

# ラベル名の設定
xlabel_name = "sepal width (cm)"
# ヒストグラム設定
df0[xlabel_name].hist(color="b",alpha=0.5)
df1[xlabel_name].hist(color="r",alpha=0.5)
df2[xlabel_name].hist(color="g",alpha=0.5)
# グラフラベル設定
plt.xlabel(xlabel_name)
# グラフの表示
plt.show()
```

```python
%matplotlib inline
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt

# sklearnでアヤメのデータセットを変数に格納
iris = datasets.load_iris()

# pandasでアヤメのデータセットをデータフレームに格納
df = pd.DataFrame(iris.data)

# データフレームのカラム名を変更
df.columns = iris.feature_names
# アヤメの品種を表示
df["target"] = iris.target

#=================================================================1
#												二つの特徴量から散布図を作成
#========================================================2021.12.19
# v1: 散布図の描画

# グラフサイズ設定
plt.figure(figsize=(5,5))

# アヤメの品種次第で格納するデータフレームを分岐
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]

# ラベル名の設定
xlabel_name = "sepal width (cm)"
ylabel_name = "sepal length (cm)"
# 散布図設定
plt.scatter(df0[xlabel_name],df0[ylabel_name],color="b",alpha=0.5)
plt.scatter(df1[xlabel_name],df1[ylabel_name],color="r",alpha=0.5)
plt.scatter(df2[xlabel_name],df2[ylabel_name],color="g",alpha=0.5)
plt.xlabel(xlabel_name)
plt.ylabel(ylabel_name)
plt.grid()
plt.show()
```
